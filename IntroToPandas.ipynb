Data Frames

```python
Data= {"Name": ["Jimmy","Karan","Krish"],
       "Age":[14,18,15],
       "Marks":[87,85,90]}
print(Data)
```

```python
# we can also add custom index in pandas dataframe
marks = pd.DataFrame(Data,index=["Student1","Student2","Student3"])
print(marks)
```

```python
marks.set_index("Name",inplace= True)
print(marks)
```

group by pandas

```python
df = pd.DataFrame({
    "Name": ["Aman", "Saurav", "Karan", "Shubham","abhishek", None, "Sameer"],
    "Age": [22,23,None, 22, 25, 26,27],
    "Marks": [85, None, 88,90, 56,72, None],
    "Subject":["MATHS","PHYSICS","MATHS","CHEMISTRY","MATHS","PHYSICS","CHEMISTRY"]
})
df
```

```python
df.groupby('Subject')['Marks'].mean()
```

```python
df.groupby('Subject')['Name'].count()
```

```python
df.groupby('Subject')['Marks'].agg(['mean','max','min'])
```

```python
df.groupby('Subject')['Marks'].agg(['mean','max','min','count','std',])
```

```python
df.groupby(['Subject','Age'])['Marks'].agg(['mean','max','min','count','std',])
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df= pd.DataFrame({
    "Name": ["Aman", "Saurav", "Karan", "Shubham","RIYA"," kunal", "Sameer"],
    "Age": [22,23,24, 22, 25, 26,27],
    "Marks": [85, 66, 88,90, 56,72, 77],
    "Gender":["MALE",'MALE','MALE',"MALE","FEMALE","MALE","MALE"],
    "Subject":["MATHS","PHYSICS","MATHS","CHEMISTRY","MATHS","PHYSICS","CHEMISTRY"]
})

print(df)
```

```python
plt.plot(df["Marks"])
plt.title("Marks Trend")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
```

```python
plt.bar(df["Name"],df["Marks"])
plt.title("Marks of students")
plt.xlabel("Student Name")
plt.ylabel("Marks")
plt.show()
```

```python
plt.hist(df["Age"],bins=5)
plt.title("Age distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.scatter(df["Age"],df["Marks"])
plt.title("Age vs Marks")
plt.xlabel("Age")
plt.ylabel("Marks")
plt.show()
```

```python
gender_counts=df["Gender"].value_counts()
plt.pie(gender_counts,labels=gender_counts.index,autopct="%1.1f%%")
plt.title("Gender Distribution")
plt.show()
```

```python
plt.bar(df["Name"],df["Marks"],color="green")
plt.title("Student Marks")
plt.xlabel("Student Name")
plt.ylabel("Marks")
plt.show()
```
