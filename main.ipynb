```python
import numpy as np

# 1D array
a = np.array([1, 2, 3, 4, 5])
print(a)

# 2D array
b = np.array([[1, 2, 3], [4, 5, 6]])
print(b)

# Basic operations
print(a + 10)
print(a * 2)
print(a.mean())
print(a.sum())

# Shape and size
print(b.shape)
print(b.size)
```
