The Scenario: TechCorp HR Data

Context: You are an HR Data Analyst for "TechCorp". You have been provided with a DataFrame named df containing employee records.

The DataFrame df has the following columns:

Name: The employee’s name (contains some missing values).

Department: The department they work in (e.g., IT, Sales, HR).

Role: Their job title (e.g., Analyst, Manager, Developer).

Experience: Years of experience.

Salary: Annual salary in USD.

Score: Most recent performance review score (contains missing values for new hires).

Note: Assume the DataFrame df is already loaded into your Python environment.

```python
df =pd.DataFrame ({
    "Name": ["Alice", "Bob", None, "David", "Eva", "Frank", "Grace", None, "Henry", "Ivy"],
    "Department": ["IT", "Sales", "HR", "IT", "Sales", "HR", "IT", "Sales", "HR", "IT"],
    "Role": ["Developer", "Manager", "Analyst", "Developer", "Executive",
             "Manager", "Analyst", "Executive", "Analyst", "Developer"],
    "Experience": [5, 8, 2, 6, 3, 10, 4, 1, 7, 9],
    "Salary": [75000, 90000, 50000, 80000, 60000, 95000, 70000, 45000, 85000, 100000],
    "Score": [4.5, 4.0, None, 4.2, None, 4.8, 3.9, None, 4.1, 4.7]
})
df


# Basic checks
print(df.shape)
print(df.dtypes)

# Missing values
print(df.isna().sum())

# Average salary per department
print(df.groupby("Department")["Salary"].mean())

# High performers
print(df[df["Score"] > 4.5])

# Employees with missing score
print(df[df["Score"].isna()])
```

Exercise Questions – Part 1

Write the exact Pandas code to answer the following questions.

Basic Averages

1. What is the average (mean) Salary for each Department?
2. What is the average (mean) Performance Score for each Role?

Counting and Sizing
3. How many actual, non-null Performance Score entries exist for each Department?
4. What is the total number of employee records (rows) in each Department, regardless of whether data is missing?

```python
# 1️⃣ Average Salary for each Department
df.groupby("Department")["Salary"].mean()

# 2️⃣ Average Performance Score for each Role
df.groupby("Role")["Score"].mean()

# 3️⃣ Non-null Performance Score count per Department
df.groupby("Department")["Score"].count()

# 4️⃣ Total employee records per Department
df.groupby("Department").size()
```




Exercise Questions – Part 2

Multiple Aggregations

1. Generate a summary table for Salary based on Department that shows the mean, maximum, minimum, and count in a single line of code.
2. Generate the same summary table (mean, max, min, count) but analyze the Performance Score based on the Role.

Grouping by Multiple Columns

3. Find the average Salary grouped by both Department and Role.
4. Find the average Performance Score grouped by both Department and Experience level.

```python
# 1️⃣ Salary summary by Department
df.groupby("Department")["Salary"].agg(["mean", "max", "min", "count"])

# 2️⃣ Performance Score summary by Role
df.groupby("Role")["Score"].agg(["mean", "max", "min", "count"])

# 3️⃣ Average Salary by Department and Role
df.groupby(["Department", "Role"])["Salary"].mean()

# 4️⃣ Average Performance Score by Department and Experience
df.groupby(["Department", "Experience"])["Score"].mean()
```

Domain: Health & Fitness

Using the following Raw Data (IDs 1–20), create the required data visualizations.

Raw Data:

Exercise Hrs:
1, 2, 3, 1, 0, 2, 3, 1, 0, 2, 1, 3, 2, 1, 0, 2, 3, 1, 2, 1

Calories Intake:
2000, 2200, 2500, 2100, 1800, 2300, 2600, 2000, 1900, 2200, 2100, 2400, 2300, 2000, 1800, 2200, 2500, 2100, 2300, 2000

Steps per Day:
5000, 7000, 9000, 6000, 4000, 8000, 10000, 5500, 4500, 7200, 6000, 8500, 9200, 5800, 4200, 7600, 9800, 6100, 8000, 5900

Heart Rate:
72, 75, 78, 74, 70, 76, 80, 73, 71, 75, 74, 77, 79, 73, 70, 76, 81, 74, 77, 74

Required Plots:

Create a Scatter Plot showing the relationship between Exercise Hours and Heart Rate.

Create a Histogram to display the distribution of Steps per Day.

Create a Bar Chart showing Calories Intake for each ID.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Raw Data
exercise = [1,2,3,1,0,2,3,1,0,2,1,3,2,1,0,2,3,1,2,1]
calories = [2000,2200,2500,2100,1800,2300,2600,2000,1900,2200,2100,2400,2300,2000,1800,2200,2500,2100,2300,2000]
steps = [5000,7000,9000,6000,4000,8000,10000,5500,4500,7200,6000,8500,9200,5800,4200,7600,9800,6100,8000,5900]
heart_rate = [72,75,78,74,70,76,80,73,71,75,74,77,79,73,70,76,81,74,77,74]

# Create DataFrame
df = pd.DataFrame({
    "Exercise_Hours": exercise,
    "Calories": calories,
    "Steps": steps,
    "Heart_Rate": heart_rate
})

print("Dataset")
print(df)
```

```python
print("\nMean")
print(df.mean())

print("\nMedian")
print(df.median())

print("\nStandard Deviation")
print(df.std())

print("\nCorrelation")
print(df.corr())
```

```python
plt.scatter(df["Exercise_Hours"], df["Heart_Rate"])
plt.title("Exercise Hours vs Heart Rate")
plt.xlabel("Exercise Hours")
plt.ylabel("Heart Rate")
plt.show()
```

```python
plt.hist(df["Steps"], bins=8)
plt.title("Distribution of Daily Steps")
plt.xlabel("Steps")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.bar(range(len(df["Calories"])), df["Calories"])
plt.title("Calories Intake")
plt.xlabel("Person ID")
plt.ylabel("Calories")
plt.show()
```

Inference:
From the scatter plot, we observe that as exercise hours increase, heart rate also tends to increase, showing a positive relationship. The histogram shows that most individuals walk between 6000–9000 steps per day. The bar chart indicates variation in daily calorie intake among individuals. Overall, the analysis suggests that physical activity such as exercise and walking contributes to maintaining an active heart rate and balanced health patterns.
